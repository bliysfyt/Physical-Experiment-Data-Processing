```python
import pandas as pd
import numpy as np
```

```python
data1=pd.read_csv("超声测速.csv")
data1
```

```python
for j in ["驻波法位置","相位法位置"]:
    for i in range(10):
        data1.loc[i,f"{j}差"]=data1.loc[i+10,j]-data1.loc[i,j]
data1
```

```python
t=29
speed_s=331.45*np.sqrt((t+273.15)/273.15)
speed_s
```

```python
differ_1_sum = np.sum(data1["驻波法位置差"])*1E-3
wavelength_1 = differ_1_sum / 50
frequency_1 = 35.092e3
speed_1 = wavelength_1 * frequency_1

differ_1_mean = differ_1_sum / 10
relative_average_deviation_1 = (np.sum(np.abs(data1["驻波法位置差"]*1E-3-differ_1_mean))/10) / differ_1_mean * 100
relative_error_1 = (speed_1-speed_s) / speed_s
print(f"{differ_1_sum=},{wavelength_1=},{frequency_1=},{speed_1=}\n{relative_average_deviation_1=}%,{relative_error_1=}%")
```

```python
differ_2_sum = np.sum(data1["相位法位置差"])*1E-3
wavelength_2 = differ_2_sum / 100
frequency_2 = 35.087e3
speed_2 = wavelength_2 * frequency_2

differ_2_mean = differ_2_sum / 10
relative_average_deviation_2 = (np.sum(np.abs(data1["相位法位置差"]*1E-3-differ_2_mean)/10)) / differ_2_mean * 100
relative_error_2 = (speed_2-speed_s) / speed_s *100
print(f"{differ_2_sum=},{wavelength_2=},{frequency_2=},{speed_2=}\n{relative_average_deviation_2=}%,{relative_error_2=}%")
```
